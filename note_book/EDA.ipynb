```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%cd C:\Users\manhm\Desktop\MobileNetV2_fish\data
```

```python


train_data = pd.read_csv('train.csv')
val_data = pd.read_csv('val.csv')
test_data = pd.read_csv('test.csv')
```

```python
print("Train data shape:", train_data.shape)
print("Validation data shape:", val_data.shape) 
print("Test data shape:", test_data.shape)
```

```python
 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
train_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax1)
ax1.set_title('Train')

val_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax2)
ax2.set_title('Validation')

test_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax3)
ax3.set_title('Test')

plt.tight_layout()
plt.show()
```

```python
train_data_aug = pd.read_csv('train_balanced.csv')
```

```python
 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
train_data_aug['score'].value_counts(normalize=True).plot(kind='bar', ax=ax1)
ax1.set_title('Train_augmented')

val_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax2)
ax2.set_title('Validation')

test_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax3)
ax3.set_title('Test')

plt.tight_layout()
plt.show()
```
